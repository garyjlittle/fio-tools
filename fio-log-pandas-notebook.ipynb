```python
#create a dataframe containing the data from each disk
grp=pd.DataFrame()
```

```python
#insert the existing columns into this new DF
grp.insert(0,"fio_bw1",bw1["value"])
grp.insert(1,"bw2",bw2["value"])
grp.insert(2,"bw3",bw3["value"])
```

```python
#append a column containing the sum of the BW across each disk
grp=grp.assign(total=grp.sum(axis=1))
```
