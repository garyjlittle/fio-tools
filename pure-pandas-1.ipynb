```python
import pandas as pd
import os
```

```python
fio_dataframe={}
fio_file_list=['2g-read-60s_bw.1.log','2g-read-60s_bw.2.log','2g-read-60s_bw.3.log']
fio_file_pathname='/Users/gary/git/fio-tools/'
```

```python
# read an fio file as a simple dataframe.
plotdf=pd.DataFrame()
i=0
for fiologname in fio_file_list:
    df=pd.read_csv(os.path.join(fio_file_pathname,fiologname))
    bw_columns=['time','bandwidth','direction','prio']
    df.columns=bw_columns
    #fio_dataframe[i]=df
    plotdf.insert(i,fiologname,df["bandwidth"])
    i=i+1
plotdf=plotdf.assign(total=plotdf.sum(axis=1))
```

```python
plotdf.plot()
```
